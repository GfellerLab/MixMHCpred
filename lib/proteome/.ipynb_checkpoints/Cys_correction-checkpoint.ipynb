```python
from collections import Counter

AA = ['A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y']


def cr_PWM(lines,PL=9,AAs=AA):
    
    PWM = [[],[]]
    for i in range(lines.shape[1]):
        PWM[0].append([])
        PWM[1].append([])


        counter = Counter(lines[:,i])
        for key in AAs:
            PWM[0][i].append(key)
            PWM[1][i].append(counter[key]/lines.shape[0])


    PWM_df = pd.DataFrame({'AA':AAs})
    for j in range(PL):
        PWM_df[str(j+1)] = PWM[1][j] 
        
    return PWM_df


def Sub_cys_corr(x,data_species,PL,AA):

    species_Alleles = list(data_species['Allele']) 

    if x in species_Alleles:
        peptides = list(data_species[(data_species['Allele'] == x) & (data_species['Ligand'].str.len()==PL)]['Ligand'])
        if len(peptides) > 100:
            lines = np.array([list(p) for p in peptides])
            return cr_PWM(lines,PL=PL,AAs=AA)
        else:
            peptides = list(data_species[(data_species['Ligand'].str.len()==PL)]['Ligand'])
            lines = np.array([list(p) for p in peptides])
            return cr_PWM(lines,PL=PL,AAs=AA)

    else:
        peptides = list(data_species[(data_species['Ligand'].str.len()==PL)]['Ligand'])
        lines = np.array([list(p) for p in peptides])
        return cr_PWM(lines,PL=PL,AAs=AA)
    

def Cys_Corr(Alleles,Alleles_H,HLA_I_non_spec,H2_I_non_spec,PL=9,AAs=AA):
    PWMs = []
    

    for x in Alleles:
        if x in Alleles_H :
            
            PWMs.append(Sub_cys_corr(x,HLA_I_non_spec,PL=PL,AA=AAs))


        else:
            PWMs.append(Sub_cys_corr(x,H2_I_non_spec,PL=PL,AA=AAs))
            
    return  PWMs

    
```
