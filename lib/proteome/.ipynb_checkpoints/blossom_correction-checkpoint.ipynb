```python
def load_dictionary(file_name):
    
    with open(file_name) as f:
        amino_acids=f.readline().strip().split('\t')
        if '\n' in amino_acids[-1]:
            amino_acids[-1]=amino_acids[-1].replace("\n",'')
    
    dictionary={}
    for i_aa,aa in enumerate(amino_acids):
        dictionary[aa]=i_aa
    return dictionary


def load_blosum(bl_path="/Users/danieltadros/Desktop/PhD/Chapter_2/code/blosum62.txt"):
    # load blosum transition matrix
    dictionary=load_dictionary(bl_path)
    trans_prob_file=open(os.path.join(bl_path))

    dict_trans=trans_prob_file.readline().strip().split('\t')
    blosum=np.zeros((len(dict_trans),len(dict_trans)))

    for i, line in enumerate(trans_prob_file):
        blosum[i,:]=[float(val) for val in line.strip().split('\t')[1:] ]
        sum_line=sum(blosum[i,:])
        #print(sum_line)
        #print(blosum[i,:])
        blosum[i,:]=[val/sum_line for val in blosum[i,:]] 
        
    return blosum
```

```python
AA=['A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y']


def Blosum_Corr(blosum_t, file_path,Alleles,PWMs,AAs=AA):
    
    PWM_c = []
    for q in range(len(Alleles)):
        peptides =pd.read_csv(file_path[q],header= None)
        N = len(peptides)
        B = 200 
        PWM = PWMs[q]
        PWM_df = pd.DataFrame({})
        T = B + N 

        for i in range(9):
            pos = []
            for j in range(20):
                x = (N * PWM[str(i+1)][j]) 
                a = 0
                for z in range(20):
                    a+= (blosum_t[j][z] * PWM[str(i+1)][z])
                x = x + (B*a)

                ps = x/T
                pos.append(ps)
            if (sum(pos)<0.999) | (sum(pos)>1.001):
                pos = pos/sum(pos)
            PWM_df[str(i+1)] = pos
        PWM_df.index = AAs
        PWM_c.append(PWM_df)    
    return PWM_c
```
